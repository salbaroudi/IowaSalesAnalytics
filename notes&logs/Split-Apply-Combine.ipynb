```python
mtcars = pd.read_csv("./data/mtcars.csv")
```

```python
#Here, the unique values of the cyl column are used as the group names. rows that have a specific cyl value are placed
#into the groups. We can see we have sub-dataframes with the cyl all the same value!
mtgroup = mtcars.groupby("cyl")
```

```python
for name, group in mtgroup:
    print(name)
    print(group)
```

```python
#we can focus our data frame on one column, and show the groups along this one column!
for name, group in mtgroup["hp"]:
    print(name)
    print(group)
```

```python
#Aggregation: For each group, we take a column, and use a summary function on the subset of the column.
mtgroup.agg(max)
```

```python
#applying multiple functions as once:
#A hiearchical column index has been produced.
mtgroup.agg([max,min])
```

```python
#Named Aggregation: Do column specific operations, pass a list of funcitons, avoid hiearchical naming!
#Note that we have no way of passing additiona arguments to the functions with this syntax
mtgroup.agg(themin=("mpg","min"),themaxx=("hp","max")) 
```

```python
#Can we apply any old function with agg?
#Indicies mismatch. Your return value has to be a single number per group!
#To properly control the output, use Named Aggregation, as above.
def fun1(x):
    #return pd.DataFrame({"a":x.sum(),"b":x.sum()})
    return x
mtgroup.agg(fun1)
```

```python
#Applying specific aggregation functions to specific columns:
#This is easier to do then NamedAggregation:
mtgroup.agg({"hp":"sum","wt":"mean"})
```

**Understanding Transformation:**

```python
#Simplest Transform: We can pick out one cell from a grouped element:

mtgroup.transform(lambda x: x.wt)
```

**Understanding Apply:**

Lets first create a reduced data frame, so we can more easily view our results

```python
mtgroup = mtcars.head(10).groupby("cyl")
```

```python
#here you can see the group passing behaviour, 
#and apply has decided to paste all the result columns together.
def g(x):
    print(x)
    return x*x

mtgroup['wt'].apply(g)
    
```

```python
#How can we reduce our dimensions?
#The return value of the applied function dictates our final output. 
#What does this look like?
def g(x):
    return x.sum()

#here, apply just concat'ed the group dataframes into one dataframe. Notice we increased our dimensions!
mtgroup['wt'].apply(g)
```

```python
#The return value of the applied function dictates our final output. 
#What does this look like?
def g(x):
    return pd.DataFrame({"p":x*x,"q":x*x}) 

#here, apply just concat'ed the group dataframes into one dataframe. Notice we increased our dimensions!
mtgroup['wt'].apply(g)
```

**Learning Point:** apply supplies the entire group series or data frame to the function
(with some kind of iterative structure to pass the cells). The output produced by the function
dictates how apply joins our group results together.

Our *agg, transform and filter* functions are just versions of apply with fixed constraints.
